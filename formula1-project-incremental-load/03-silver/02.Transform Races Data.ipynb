```python

bronze_table=f"{catalog_name}.{bronze_schema}.races"
silver_table=f"{catalog_name}.{silver_schema}.races"
```

```python
from pyspark.sql import functions as F
```

```python

races_df=(spark.table(bronze_table)).filter(F.col("batch_id")==v_batch_id)
```

```python
races_selected_df=races_df.select(
    F.col("season"),
    F.col("round"),
    F.col("raceName"),
    F.col("date"),
    F.col("circuitId"),
    F.col("ingestion_timestamp"),
    F.col("source_file"),
    F.col('batch_id')
)
```

```python
races_renamed_df=(
    races_selected_df
    .withColumnsRenamed({"circuitId":"circuit_id",
                         "raceName":"race_name",
                         "date":"race_date"})
)
```

```python
races_distinct_df=races_renamed_df.dropDuplicates(["season","round"])
```

```python
races_final_df=(races_distinct_df
    .withColumn('race_name',F.initcap(F.col("race_name"))))
```

```python
write_to_silver(
    input_df=races_final_df,
    target_table=silver_table,
    merge_condition="t.season = s.season AND t.round = s.round",
    columns_to_update=["season","round","race_name","date","circuit_id","ingestion_timestamp","source_file","batch_id"] 
)
```
