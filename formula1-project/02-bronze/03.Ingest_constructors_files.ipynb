```python
source_file=f"{landing_folder_path}/constructors.json"
table_name=f"{catalog_name}.{bronze_schema}.constructors"
```

```python
constructor_schema = """
constructorId STRING,
name STRING,
nationality STRING,
url STRING
"""
```

```python
constructors_df=(
    spark.read
    .format('json')
    .option('mode','FAILFAST')
    .schema(constructor_schema)
    .load(source_file)
)
```
